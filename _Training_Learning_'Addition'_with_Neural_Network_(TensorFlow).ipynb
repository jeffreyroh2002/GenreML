```python
# Objective: using a neural network to learn how to add
# ex)    input = array([ [0.1, 0.2] , [0.2, 0.2] ])
#        output = array ([ [0.3] [0.4] ])
```

```python
import numpy as np
from random import random
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
#Function that Generates Dataset

def generate_dataset(num_samples, test_size):
  #randomly generate num_samples of x,y pairs
  x = np.array([[random()/2 for _ in range(2)] for _ in range(num_samples)])  
  y = np.array([[i[0] + i[1]] for i in x])   # y = answer
  x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = test_size)  #if test_size 0.3, 30% of data used for testing
  return x_train, x_test, y_train, y_test

if __name__ == "__main__":
  x_train, x_test, y_train, y_test = generate_dataset(5000, 0.3)
  print("x_test: {}".format(x_test))
  print("y_test: {}".format(y_test))

  #building our model:  2 --> 5 --> 1
  #keras allows us to use complex neural networks simply

  model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, input_dim=2, activation="sigmoid"),   #dense connects all layers together (fully connected)
    tf.keras.layers.Dense(1, activation="sigmoid")
  ])
  

  optimiser = tf.keras.optimizers.SGD(learning_rate = 0.1)  #stocastic gradient descent
  model.compile(optimizer=optimiser, loss="MSE")    #error function 

  #train model  -->  does both front propogation and back propogation
  model.fit(x_train, y_train, epochs = 100)

  #evaluate model
  print("Model evaluation:")
  model.evaluate(x_test, y_test, verbose = 1)

  #making predictions
  data = np.array([[0.1, 0.2], [0.2, 0.2]])
  predictions = model.predict(data)
  
  print("Predictions:")
  for d, p in zip(data, predictions):
    print("{} + {} = {}".format(d[0], d[1], p[0]))

```
